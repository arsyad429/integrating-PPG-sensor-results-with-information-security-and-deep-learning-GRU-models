```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
PATH = r"e:\PPGSensorProject\streamlit_dashboard_monitoring\captured_data"
```

```python
baseline_data_folder_path = os.path.join(PATH, "base_version")
ascon_data_folder_path = os.path.join(PATH, "with_ASCON")
sha3_data_folder_path = os.path.join(PATH, "with_SHA3")
ascon_sha3_folder_path = os.path.join(PATH, "with_ASCON_SHA3")
```

```python
base_files = os.listdir(baseline_data_folder_path)
ascon_files = os.listdir(ascon_data_folder_path)
sha3_files = os.listdir(sha3_data_folder_path)
ascon_sha3_files = os.listdir(ascon_sha3_folder_path)
```

```python
df_base = pd.DataFrame()
df_ascon = pd.DataFrame()
df_sha3 = pd.DataFrame()
df_sha3_ascon = pd.DataFrame()
```

```python
for file in base_files:
    df_file = pd.read_excel(os.path.join(baseline_data_folder_path, file))
    df_base = pd.concat([df_base, df_file])
df_base
```

```python
for file in ascon_files:
    df_file = pd.read_excel(os.path.join(ascon_data_folder_path, file))
    df_ascon = pd.concat([df_ascon, df_file])
df_ascon
```

```python
for file in sha3_files:
    df_file = pd.read_excel(os.path.join(sha3_data_folder_path, file))
    df_sha3 = pd.concat([df_sha3, df_file])
df_sha3
```

```python
for file in ascon_sha3_files:
    df_file = pd.read_excel(os.path.join(ascon_sha3_folder_path, file))
    df_sha3_ascon = pd.concat([df_sha3_ascon, df_file])
df_sha3_ascon
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# ---------------------------------------------------------
# 1. PERSIAPAN DATA
# ---------------------------------------------------------
# Pastikan kolom Timestamp dibaca sebagai format waktu (datetime)
df_base['Timestamp_PC'] = pd.to_datetime(df_base['Timestamp_PC'])

# Mengurutkan data berdasarkan waktu untuk memastikan grafik tidak berantakan
df_base = df_base.sort_values('Timestamp_PC').reset_index(drop=True)

# ---------------------------------------------------------
# 2. GRAFIK 1: Perbandingan Sinyal Mentah vs Filtered
# ---------------------------------------------------------
# Menggunakan plotly.graph_objects (go) agar lebih fleksibel mem-plot 2 sumbu y
fig_ppg = go.Figure()

# Garis 1: Sinyal Mentah (Sumbu Y Utama - Kiri)
fig_ppg.add_trace(go.Scatter(
    x=df_base['Timestamp_PC'], 
    y=df_base['PPG_Raw'],
    name='Sinyal Mentah (Ada DC Offset)',
    mode='lines',
    line=dict(color='#FFA500', width=1) # Warna oranye
))

# Garis 2: Sinyal Filtered (Sumbu Y Kedua - Kanan)
fig_ppg.add_trace(go.Scatter(
    x=df_base['Timestamp_PC'], 
    y=df_base['PPG_Filtered'],
    name='Sinyal Bersih (DC Removal)',
    mode='lines',
    line=dict(color='#00FF7F', width=2), # Warna hijau terang
    yaxis='y2'
))

fig_ppg.update_layout(
    title='1. Perbandingan Sinyal PPG: Mentah vs Filtered (DC Removal)',
    template='plotly_dark',
    yaxis=dict(title='Nilai IR Mentah (ADC)'),
    yaxis2=dict(title='Amplitudo Nadi Bersih', overlaying='y', side='right'),
    legend=dict(x=0, y=1.1, orientation='h')
)
fig_ppg.show()

# ---------------------------------------------------------
# 3. GRAFIK 2: Stabilitas Koneksi MQTT (Latency)
# ---------------------------------------------------------
fig_lat = px.line(
    df_base, 
    x='Timestamp_PC', 
    y='Latency_ms', 
    title='2. Stabilitas Koneksi MQTT (Latency Jaringan)',
    template='plotly_dark'
)
# Menambahkan garis rata-rata latency
avg_lat = df_base['Latency_ms'].mean()
fig_lat.add_hline(y=avg_lat, line_dash="dash", line_color="red", 
                  annotation_text=f"Rata-rata: {avg_lat:.1f} ms")

fig_lat.update_traces(line_color='#00BFFF', line_width=1.5) # Ubah warna sedikit agar beda
fig_lat.update_layout(yaxis_title='Waktu Tempuh (ms)')
fig_lat.show()

# ---------------------------------------------------------
# 4. GRAFIK 3: Throughput Pengiriman Data
# ---------------------------------------------------------
# Menghitung berapa banyak pesan yang masuk di setiap detiknya
df_throughput = df_base.groupby(df_base['Timestamp_PC'].dt.floor('s')).size().reset_index(name='Message_Count')

# Asumsi 1 pesan JSON mentah (sebelum ditambah ASCON/SHA3) besarnya sekitar ~120 Bytes
ESTIMASI_PAYLOAD_BYTES = 120
df_throughput['Throughput_Bytes_per_Sec'] = df_throughput['Message_Count'] * ESTIMASI_PAYLOAD_BYTES

fig_thr = go.Figure()

# Plot Jumlah Pesan per Detik
fig_thr.add_trace(go.Bar(
    x=df_throughput['Timestamp_PC'], 
    y=df_throughput['Message_Count'],
    name='Pesan/Detik',
    marker_color='#1E90FF'
))

# Plot Estimasi Throughput (Bytes/Detik) di sumbu Y kedua (sebelah kanan)
fig_thr.add_trace(go.Scatter(
    x=df_throughput['Timestamp_PC'], 
    y=df_throughput['Throughput_Bytes_per_Sec'],
    name='Estimasi Throughput (Bytes/s)',
    mode='lines+markers',
    line=dict(color='#FF69B4', width=2),
    yaxis='y2'
))

fig_thr.update_layout(
    title='3. Throughput Data Awal (Messages/sec & Bytes/sec)',
    template='plotly_dark',
    yaxis=dict(title='Jumlah Pesan per Detik'),
    yaxis2=dict(title='Throughput (Bytes/sec)', overlaying='y', side='right'),
    legend=dict(x=0, y=1.1, orientation='h')
)
fig_thr.show()
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# ---------------------------------------------------------
# 1. PERSIAPAN DATA
# ---------------------------------------------------------
df_ascon['Timestamp_PC'] = pd.to_datetime(df_ascon['Timestamp_PC'])
df_ascon = df_ascon.sort_values('Timestamp_PC').reset_index(drop=True)

# ---------------------------------------------------------
# 2. GRAFIK 1: Distribusi Waktu Enkripsi di ESP32
# ---------------------------------------------------------
# Menggunakan Histogram + Boxplot untuk melihat seberapa konsisten kecepatan ASCON
fig_enc = px.histogram(
    df_ascon, 
    x='Enc_Time_us', 
    title='1. Distribusi Waktu Enkripsi ASCON-128 di ESP32 (Microseconds)',
    template='plotly_dark',
    color_discrete_sequence=['#00FF7F'],
    marginal='box' # Menambahkan boxplot di atas histogram
)
fig_enc.update_layout(
    xaxis_title='Waktu Enkripsi (μs)', 
    yaxis_title='Jumlah Kemunculan (Frekuensi)'
)
fig_enc.show()

# ---------------------------------------------------------
# 3. GRAFIK 2: Tren Waktu Dekripsi di Komputer (PC)
# ---------------------------------------------------------
fig_dec = px.line(
    df_ascon, 
    x='Timestamp_PC', 
    y='Dec_Time_ms', 
    title='2. Waktu Dekripsi ASCON-128 di Python/PC (Milliseconds)',
    template='plotly_dark'
)
avg_dec = df_ascon['Dec_Time_ms'].mean()
fig_dec.add_hline(y=avg_dec, line_dash="dash", line_color="#FF69B4", 
                  annotation_text=f"Rata-rata: {avg_dec:.3f} ms")
fig_dec.update_traces(line_color='#1E90FF', line_width=1.5)
fig_dec.update_layout(yaxis_title='Waktu Dekripsi (ms)')
fig_dec.show()

# ---------------------------------------------------------
# 4. GRAFIK 3: Ilustrasi Perubahan Ukuran Ciphertext (Overhead Data)
# ---------------------------------------------------------
# ASCON-128 tidak menggunakan padding (ukuran plaintext = ciphertext)
# TETAPI ASCON menambahkan 16 Bytes Authentication Tag (MAC).
ESTIMASI_PLAINTEXT_BYTES = 120
ASCON_TAG_BYTES = 16

fig_size = go.Figure(data=[
    go.Bar(name='Plaintext (Data Asli)', x=['Ukuran Paket Data'], y=[ESTIMASI_PLAINTEXT_BYTES], marker_color='#1E90FF'),
    go.Bar(name='ASCON Authentication Tag (Overhead)', x=['Ukuran Paket Data'], y=[ASCON_TAG_BYTES], marker_color='#FF4500')
])

fig_size.update_layout(
    barmode='stack', 
    title='3. Perubahan Ukuran Data (Plaintext vs Ciphertext ASCON-128)',
    template='plotly_dark',
    yaxis_title='Ukuran (Bytes)'
)
fig_size.show()
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# ---------------------------------------------------------
# 1. PERSIAPAN DATA ESTIMASI PAYLOAD
# ---------------------------------------------------------
# Estimasi ukuran maksimal JSON payload (dalam Bytes/Karakter) 
# berdasarkan struktur StaticJsonDocument di ESP32 Anda.

data_payload = {
    'Arsitektur Keamanan': [
        '1. Baseline (Plaintext)', 
        '2. SHA3-256 (Integrity Only)', 
        '3. ASCON-128 (Confidentiality)', 
        '4. Level 9 (ASCON + SHA3)'
    ],
    'Ukuran_Payload_Bytes': [130, 250, 350, 490],
    'Komponen_Tambahan': [
        'Hanya Data Medis & Telemetri',
        '+ Hash 64 Karakter & Waktu Hash',
        '+ Hex Ciphertext (x2), Tag 16B, Nonce 32 Karakter',
        '+ Hex Ciphertext (x2), Tag 16B, Nonce, & Hash 64 Karakter'
    ]
}

df_payload = pd.DataFrame(data_payload)

# ---------------------------------------------------------
# 2. VISUALISASI BAR CHART
# ---------------------------------------------------------
fig = px.bar(
    df_payload, 
    x='Arsitektur Keamanan', 
    y='Ukuran_Payload_Bytes',
    text='Ukuran_Payload_Bytes',
    color='Arsitektur Keamanan',
    color_discrete_sequence=['#00FF7F', '#1E90FF', '#FFA500', '#FF4500'],
    hover_data=['Komponen_Tambahan'],
    title='📊 Perbandingan Ukuran Maksimal Payload MQTT (Bytes)',
    template='plotly_dark'
)

# Menyesuaikan posisi teks dan tampilan layout
fig.update_traces(
    textposition='outside', 
    textfont_size=14,
    marker_line_color='white',
    marker_line_width=1.5
)

fig.update_layout(
    yaxis_title='Ukuran Payload (Bytes)',
    xaxis_title='Skenario Sistem',
    yaxis=dict(range=[0, 600]), # Memberikan ruang kosong di atas agar teks angka tidak terpotong
    showlegend=False
)

fig.show()
```

```python
import pandas as pd
import plotly.express as px

# ---------------------------------------------------------
# 1. PERSIAPAN DATA LATENSI MAKSIMAL
# ---------------------------------------------------------
# Mengambil nilai tertinggi (max) dari kolom Latency_ms di tiap dataframe
data_max_latency = {
    'Arsitektur Keamanan': [
        '1. Baseline (Plaintext)', 
        '2. SHA3-256 (Integrity Only)', 
        '3. ASCON-128 (Confidentiality)', 
        '4. Level 9 (ASCON + SHA3)'
    ],
    'Max_Latency_ms': [
        df_base['Latency_ms'].max(),
        df_sha3['Latency_ms'].max(),
        df_ascon['Latency_ms'].max(),
        df_sha3_ascon['Latency_ms'].max()
    ]
}

df_max_lat = pd.DataFrame(data_max_latency)

# ---------------------------------------------------------
# 2. VISUALISASI BAR CHART
# ---------------------------------------------------------
fig_max_lat = px.bar(
    df_max_lat, 
    x='Arsitektur Keamanan', 
    y='Max_Latency_ms',
    text='Max_Latency_ms',
    color='Arsitektur Keamanan',
    color_discrete_sequence=['#00FF7F', '#1E90FF', '#FFA500', '#FF4500'],
    title='📶 Perbandingan Waktu Latensi Terlama (Maximum Latency) MQTT',
    template='plotly_dark'
)

# Menyesuaikan posisi teks dan tampilan garis luar bar
fig_max_lat.update_traces(
    textposition='outside', 
    texttemplate='%{text} ms', # Menambahkan satuan ms di atas bar
    textfont_size=14,
    marker_line_color='white',
    marker_line_width=1.5
)

# Mengatur skala sumbu Y agar teks di puncaknya tidak terpotong
max_y_value = df_max_lat['Max_Latency_ms'].max()
fig_max_lat.update_layout(
    yaxis_title='Latensi Maksimal (ms)',
    xaxis_title='Skenario Sistem',
    yaxis=dict(range=[0, max_y_value * 1.2]), # Memberikan sisa ruang 20% di atas
    showlegend=False
)

fig_max_lat.show()
```
